```python
import pandas as pd
import numpy as np
```

```python
data_name = 'Cascode'
data = pd.read_csv(f'data/{data_name}/{data_name}.csv')
idx = 4

x = np.asarray(data.iloc[:,0:idx])
np.save(f'data/{data_name}/x.npy', x)

y = np.asarray(data.iloc[:,idx:])
np.save(f'data/{data_name}/y.npy', y)
```

```python
data_name = 'Cascode'

# test_df = pd.DataFrame(columns=['C0', 'C1', 'WNOUT', 'WPOUT', 'WCS', 'WN1', 'WP1', 'Bandwidth', 'Power Consumption', 'Voltage Gain']) # two_stage
# test_df = pd.DataFrame(columns=['C1', 'C2', 'Ld', 'Lg', 'Ls', 'WN1', 'WN2', 'GT', 'S11', 'NF', 'BW']) # LNA
# test_df = pd.DataFrame(columns=['VDD', 'Vgate', 'Wn', 'Rd', 'Bandwidth', 'Power Consumption', 'Voltage Gain']) # SingleStageAmplifier
test_df = pd.DataFrame(columns=['Wbias', 'Rd', 'Wn1', 'Wn2', 'Bandwidth', 'Power Consumption', 'Voltage Gain']) # Cascode

test_x = np.load(f'data/{data_name}/test_x.npy')
test_y = np.load(f'data/{data_name}/test_y.npy')

for i in range(test_x.shape[0]):
    test_df.loc[i] = list(test_x[i]) + list(test_y[i])

test_df.to_csv(f'data/{data_name}/test_{data_name}.csv', index=False)
test_df
```
